```python
#delete operation by value in dobulylinkedlist
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    def insert_at_end(self, data):
        newnode = Node(data)
        if self.head is None:
            self.head = newnode
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = newnode
        newnode.prev = temp
    def delete_by_value(self, value):
        if self.head is None:
            print("List is empty.")
            return
        temp = self.head
        if temp.data == value:
            self.head = temp.next
            if self.head:
                self.head.prev = None
            temp = None
            return
        while temp:
            if temp.data == value:
                break
            temp = temp.next
        if temp is None:
            print(f"Value {value} not found in the list.")
            return
        if temp.next:
            temp.next.prev = temp.prev
        if temp.prev:
            temp.prev.next = temp.next
        temp = None
    def display(self):
        temp = self.head
        print("DoublyLinkedList:")
        while temp:
            print(temp.data, end=' <--> ')
            temp = temp.next
        print("None")
dll = DoublyLinkedList()
n = int(input("Enter the number of elements to insert at end: "))
for i in range(n):
    val = int(input(f"Enter element {i+1}: "))
    dll.insert_at_end(val)
dll.display()
value_to_delete = int(input("Enter value to delete: "))
dll.delete_by_value(value_to_delete)
dll.display()
```
